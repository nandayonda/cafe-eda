```python
#for page 2 onwards
def get_cafe_home(link):
    cafe_data = requests.get(link)
    cafe_soup = BeautifulSoup(cafe_data.content, "html.parser")
    cafe_container = cafe_soup.find(id="restaurant_contents")
    cafe_lists = cafe_container.find_all("div", class_="restaurant-result-wrapper best-rating")

    return cafe_lists

def get_cafe_link(cafe_list):
    cafe_links = []
    for cafe in cafe_list:
        cafe_content = {}
        cafe = cafe.find("div", class_="item-info")
        cafe_content["name"] = cafe.find("h3", class_="item-name").text.strip()
        cafe_content["page"] = base_url + cafe.find("a")["href"]
        cafe_links.append(cafe_content)

    return cafe_links
```

```python
#scrape first page
cafe_data = requests.get("https://pergikuliner.com/restaurants?utf8=%E2%9C%93&search_place=&default_search=Bandung&search_name_cuisine=&commit=")

cafe_soup = BeautifulSoup(cafe_data.content, "html.parser")
cafe_container = cafe_soup.find(id="restaurant_contents")

cafe_lists = cafe_container.find_all("div", class_="restaurant-result-wrapper best-rating")
next_page = cafe_container.find("div", class_="result-footer").find("a", id="next")["href"]

cafe_links = []
for cafe in cafe_lists:
    cafe_content = {}
    cafe = cafe.find("div", class_="item-info")
    cafe_content["name"] = cafe.find("h3", class_="item-name").text.strip()
    cafe_content["page"] = base_url + cafe.find("a")["href"]
    cafe_links.append(cafe_content)

for i in range(2,126):
    cafe = get_cafe_home(f"https://pergikuliner.com/restaurants?default_search=Bandung&page={i}&search_name_cuisine=&search_place=")
    link_cafe = get_cafe_link(cafe)
    cafe_links.extend(link_cafe)
    time.sleep(5)
```

```python
import json
with open("cafe_data.json", "w") as file:
    json.dump(cafe_links, file)
```
